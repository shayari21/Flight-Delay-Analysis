```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Define pipeline steps
steps = [
    "1. Data Preprocessing", 
    "2. Data Splitting", 
    "3. Model Pipeline", 
    "4. Model Training & Prediction", 
    "5. Model Evaluation", 
    "6. Hyperparameter Tuning (Optional)", 
    "7. Deployment & Monitoring(Optional)"
]

# Add nodes to the graph
for step in steps:
    G.add_node(step)

# Define connections (flow)
edges = [
    ("1. Data Preprocessing", "2. Data Splitting"),
    ("2. Data Splitting", "3. Model Pipeline"),
    ("3. Model Pipeline", "4. Model Training & Prediction"),
    ("4. Model Training & Prediction", "5. Model Evaluation"),
    ("5. Model Evaluation", "6. Hyperparameter Tuning (Optional)"),
    ("6. Hyperparameter Tuning (Optional)", "4. Model Training & Prediction"),
    ("5. Model Evaluation", "7. Deployment & Monitoring")
]

# Add edges to the graph
G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=42)  # Layout for visualization
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=3000, font_size=10, font_weight='bold')

# Show the diagram
plt.title("Machine Learning Pipeline for Flight Delay Prediction", fontsize=12, fontweight='bold')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph for the pipeline
G = nx.DiGraph()

# Define pipeline steps (short for PowerPoint slide)
steps = {
    "1. Data Preprocessing": "Handle Missing Values\nEncode Categorical Features\nScale Numerical Data",
    "2.Data Splitting": "Train-Test Split (70%-30%)",
    "3. Model Pipeline": "Feature Scaling  StandardScaler\nModel Selection  RandomForest, XGBoost, LSTM",
    "4. Model Training & Prediction": "Train Model\nMake Predictions",
    "5.Model Evaluation": "Metrics: MAE, RMSE, R\nActual vs Predicted Plot",
}

# Add nodes and edges
for step, desc in steps.items():
    G.add_node(step, label=f"{step}\n{desc}")

edges = list(zip(list(steps.keys())[:-1], list(steps.keys())[1:]))
G.add_edges_from(edges)

# Draw the graph
plt.figure(figsize=(10, 5))
pos = nx.spring_layout(G, seed=42)  # Layout for visualization

# Draw nodes with labels
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=3500, font_size=9, font_weight='bold')

# Show the title
plt.title("ML Pipeline for Flight Delay Prediction", fontsize=12, fontweight='bold')

# Show the diagram
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph for the pipeline
G = nx.DiGraph()

# Define pipeline steps (short for PowerPoint slide)
steps = {
    "1. Data Preprocessing": "Handle Missing Values\nEncode Categorical Features\nScale Numerical Data",
    "2. Data Splitting": "Train-Test Split (70%-30%)",
    "3. Model Pipeline": "Feature Scaling  StandardScaler\nModel Selection  RandomForest, XGBoost, LSTM",
    "4. Model Training & Prediction": "Train Model\nMake Predictions",
    "5. Model Evaluation": "Metrics: MAE, RMSE, R\nActual vs Predicted Plot",
}

# Add nodes and edges
for step, desc in steps.items():
    G.add_node(step, label=f"{step}\n{desc}")

edges = list(zip(list(steps.keys())[:-1], list(steps.keys())[1:]))
G.add_edges_from(edges)

# Define straight-line layout
pos = {step: (i, 0) for i, step in enumerate(steps.keys())}

# Draw the graph
plt.figure(figsize=(12, 2))
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=3500, font_size=9, font_weight='bold')

# Show the title
plt.title("ML Pipeline for Flight Delay Prediction", fontsize=12, fontweight='bold')

# Show the diagram
plt.show()
```
